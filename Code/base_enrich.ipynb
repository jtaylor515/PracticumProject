```python
import pandas as pd
import numpy as np
```

```python
df_business = pd.read_csv('/Users/jeffreytaylor/Documents/Emory/CS370/PracticumProject/Data/yelp_dataset/business.csv')
df_review = pd.read_csv('/Users/jeffreytaylor/Documents/Emory/CS370/PracticumProject/Data/yelp_dataset/review.csv')
```

```python
# Merge DataFrames
df_merged = df_business.merge(df_review, on='business_id')

# Filter for Philadelphia, food or restaurant categories (using regular expressions for broader matching)
df_filtered = df_merged[(df_merged['city'] == 'Philadelphia') & (df_merged['categories'].notna()) & (df_merged['categories'].str.contains('food|restaurant', case=False))]
```

```python
# Print length of filtered DataFrame
print("Number of reviews:", len(df_filtered))

print("Unique restaurants:",  df_filtered['business_id'].nunique())

average_words = df_filtered['text'].str.split().apply(len).mean()
print("Average number of words:", average_words)

## RANDOM SELECT

# Set the random seed for reproducibility
np.random.seed(42)

# Group the dataframe by 'business_id' and sample 10% of the reviews from each group
df_filtered_sampled = df_filtered.groupby('business_id').apply(lambda x: x.sample(frac=0.1)).reset_index(drop=True)

# Print length of filtered DataFrame
print("Number of reviews:", len(df_filtered_sampled))

print("Unique restaurants:",  df_filtered_sampled['business_id'].nunique())

average_words = df_filtered_sampled['text'].str.split().apply(len).mean()
print("Average number of words:", average_words)

df_filtered_sampled.head(10)

# save the sampled data to a new csv file
# df_filtered_sampled.to_csv('/Users/jeffreytaylor/Downloads/philadelphia_restaurant_reviews_sampled.csv', index=False)
```
